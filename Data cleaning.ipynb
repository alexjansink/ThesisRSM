```python
# Sets libiraries and options
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# reads data
df = pd.read_csv("DATA.csv")
display(df)
```

```python
#Filters data based on a few things
df = df[df["DistributionChannel"] == "anonymous"] #gets the data from published surveys

display(df)
```

```python
print(df.columns)
```

```python
#gets response data from users
df =df[['ResponseId', 'Q44', 'SC10', 'SC12', 'SC13',
       'SC14', 'SC16', 'SC18', 'SC20', 'SC21', 'SC22', 'SC28', 'SC1', 'SC3',
       'SC6', 'SC7', 'SC8', 'SC9', 'SC15', 'SC19', 'SC23', 'SC29', 'SC2',
       'SC4', 'SC5', 'SC11', 'SC17', 'SC24', 'SC25', 'SC26', 'SC27', 'SC30',
       'Q5', 'Q6', 'Q6_4_TEXT', 'Q7']]
```

```python
#converts data to long format
df["id"] = df.index

dfwide = pd.wide_to_long(df, stubnames = "SC" , i = "id", j = "number")

display(dfwide)
```

```python
# Removes the blocks of the survey people didn't react to
dfwide = dfwide.dropna(subset=["SC"])
display(dfwide)
```

```python
print(dfwide)
```

```python
#Reads the trade offs they needed to make
scen = pd.read_excel("scenario sheet.xlsx")
display(scen)
```

```python
#Adds the trade of next to the answers
dfbig= dfwide.merge(scen, left_on='number', right_on='Choice situation')
display(dfbig)
```

```python
# Makes some of the text more understandandable
dfbig = dfbig.rename(columns={"Q44": "Dummy", "Q5": "Age", "Q6": "Gender", "Q7": "Education", "ResponseId": "ID"})
display(dfbig)
```

```python
# Sorts the responses by ID because the apollo library needs it like that
dfbig = dfbig.sort_values(by=["ID"])
display(dfbig)
```

```python
# Old code before the effect coding was implemented
# dfbig.to_csv("Long_Survey", index=False)
```

```python
# Adds dummies in the k-1 format.
dfbig = pd.get_dummies(dfbig, columns= ["option1.method"],prefix=["option1.method.effect"], drop_first=True)
dfbig = pd.get_dummies(dfbig, columns= ["option2.method"],prefix=["option2.method.effect"], drop_first=True)
dfbig = pd.get_dummies(dfbig, columns= ["option1.trust"],prefix=["option1.trust.effect"], drop_first=True)
dfbig = pd.get_dummies(dfbig, columns= ["option2.trust"],prefix=["option2.trust.effect"], drop_first=True)
display(dfbig)
```

```python
# This drops for now only a entry that is not believeable old of age did the survey in 3 minutes and preffers to call itself king. This likely means that this person did not react to this survey in a reasonable way
dfbig = dfbig[dfbig["Q6_4_TEXT"].isna()]
```

```python
print(dfbig.columns)
```

```python
# Sets the true false results to 0 and 1
dfbig[['option1.method.effect_1', 'option1.method.effect_2', 'option1.method.effect_3', 'option1.method.effect_4', 'option2.method.effect_1', 'option2.method.effect_2','option2.method.effect_3', 'option2.method.effect_4', 'option1.trust.effect_1', 'option1.trust.effect_2','option2.trust.effect_1', 'option2.trust.effect_2']] = dfbig[['option1.method.effect_1', 'option1.method.effect_2', 'option1.method.effect_3', 'option1.method.effect_4', 'option2.method.effect_1', 'option2.method.effect_2','option2.method.effect_3', 'option2.method.effect_4','option1.trust.effect_1', 'option1.trust.effect_2','option2.trust.effect_1', 'option2.trust.effect_2']].astype(int)
display(dfbig)
```

```python
#display(((dfbig["option1.method.effect_1"] == 0) & (dfbig["option1.method.effect_2"] == 0)& (dfbig["option1.method.effect_3"] == 0)))
```

```python
# selects the reference group for the effects
series1 = ((dfbig["option1.method.effect_1"] == 0) & (dfbig["option1.method.effect_2"] == 0)& (dfbig["option1.method.effect_3"] == 0) & (dfbig["option1.method.effect_4"] == 0))

series2 = ((dfbig["option2.method.effect_1"] == 0) & (dfbig["option2.method.effect_2"] == 0) & (dfbig["option2.method.effect_3"] == 0) & (dfbig["option2.method.effect_4"] == 0))

series3 = ((dfbig["option1.trust.effect_1"] == 0) & (dfbig["option1.trust.effect_2"] == 0))

series4 = ((dfbig["option2.trust.effect_1"] == 0) & (dfbig["option2.trust.effect_2"] == 0))
```

```python
#checking if the data set is balanced
print(sum(series1))
print(sum(dfbig["option1.method.effect_1"]))
print(sum(dfbig["option1.method.effect_2"]))
print(sum(dfbig["option1.method.effect_3"]))
print(sum(dfbig["option1.method.effect_4"]))

print(sum(series2))
print(sum(dfbig["option2.method.effect_1"]))
print(sum(dfbig["option2.method.effect_2"]))
print(sum(dfbig["option2.method.effect_3"]))
print(sum(dfbig["option2.method.effect_4"]))
```

```python
# Calculates the amount of times an advice option is presented
meffect0 = sum(series1) + sum(series2)
meffect1 = sum(dfbig["option1.method.effect_1"]) + sum(dfbig["option2.method.effect_1"])
meffect2 = sum(dfbig["option1.method.effect_2"]) + sum(dfbig["option2.method.effect_2"])
meffect3 = sum(dfbig["option1.method.effect_3"]) + sum(dfbig["option2.method.effect_3"])
meffect4 = sum(dfbig["option1.method.effect_4"]) + sum(dfbig["option2.method.effect_4"])

# Calaculates the amount of times an opinion option is presented
opeffect0 = sum(series3) + sum(series4)
opeffect1 = sum(dfbig["option1.trust.effect_1"]) + sum(dfbig["option2.trust.effect_1"])
opeffect2 = sum(dfbig["option1.trust.effect_2"]) + sum(dfbig["option2.trust.effect_2"])
```

```python
# Does the effect coding for option 1 method
dfbig.loc[series1, "option1.method.effect_1"] = -1
dfbig.loc[series1, "option1.method.effect_2"] = -1
dfbig.loc[series1, "option1.method.effect_3"] = -1
dfbig.loc[series1, "option1.method.effect_4"] = -1

# Does the effect coding for option 2 method
dfbig.loc[series2, "option2.method.effect_1"] = -1
dfbig.loc[series2, "option2.method.effect_2"] = -1
dfbig.loc[series2, "option1.method.effect_3"] = -1
dfbig.loc[series2, "option1.method.effect_4"] = -1

# Does the effect coding for option 1 trust
dfbig.loc[series3, "option1.trust.effect_1"] = -1
dfbig.loc[series3, "option1.trust.effect_2"] = -1


# Does the effect coding for option 2 trust
dfbig.loc[series4, "option2.trust.effect_1"] = -1
dfbig.loc[series4, "option2.trust.effect_2"] = -1


```

```python
display(dfbig)
```

```python
#unweighted effect coding
dfbig.to_csv("Long_Effect", index=False)
```

```python
#weighted effect coding start
# reference paper https://journal.r-project.org/archive/2017/RJ-2017-017/RJ-2017-017.pdf
# Does the weighted effect coding for option 1 method
dfbig.loc[series1, "option1.method.effect_1"] = (meffect1/meffect0)*-1
dfbig.loc[series1, "option1.method.effect_2"] = (meffect2/meffect0)*-1
dfbig.loc[series1, "option1.method.effect_3"] = (meffect3/meffect0)*-1
dfbig.loc[series1, "option1.method.effect_4"] = (meffect4/meffect0)*-1

# Does the effect coding for option 2 method
dfbig.loc[series2, "option2.method.effect_1"] = (meffect1/meffect0)*-1
dfbig.loc[series2, "option2.method.effect_2"] = (meffect2/meffect0)*-1
dfbig.loc[series2, "option1.method.effect_3"] = (meffect3/meffect0)*-1
dfbig.loc[series2, "option1.method.effect_4"] = (meffect4/meffect0)*-1

# Does the effect coding for option 1 trust
dfbig.loc[series3, "option1.trust.effect_1"] = (opeffect1/opeffect0)*-1
dfbig.loc[series3, "option1.trust.effect_2"] = (opeffect2/opeffect0)*-1


# Does the effect coding for option 2 trust
dfbig.loc[series4, "option2.trust.effect_1"] = (opeffect1/opeffect0)*-1
dfbig.loc[series4, "option2.trust.effect_2"] = (opeffect2/opeffect0)*-1

```

```python
display(dfbig)
```
